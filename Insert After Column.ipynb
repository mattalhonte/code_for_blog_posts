```python
import pandas as pd
from typing import Callable
```

```python
cols = ['instnm', 'relaffil', 'satvrmid', 'satmtmid', 'distanceonly', 'ugds']
college = pd.read_csv('data/college.csv', index_col='instnm', usecols=cols)
```

```python
def insert_after(df: pd.DataFrame, 
                fn: Callable,
                col: str,
                after: str) -> pd.DataFrame:
    
    df = df.copy()
    
    new_col = fn(df)
    idx = df.columns.get_loc(after)
    df.insert(idx + 1, col, new_col)
    
    return df
```

```python
college.pipe(insert_after, 
             lambda x: x['satvrmid'] + x['satmtmid'],
            "test_col",
            "relaffil")
```
